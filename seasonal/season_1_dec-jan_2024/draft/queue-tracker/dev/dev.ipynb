```python
from pytimeparse import parse
import datetime

s = "5s"
total_seconds = parse(s)   # parse returns the total seconds as an integer
delta = datetime.timedelta(seconds=total_seconds)
delta = datetime.timedelta(seconds=total_seconds)
print(delta)  # prints "0:00:05"
```
